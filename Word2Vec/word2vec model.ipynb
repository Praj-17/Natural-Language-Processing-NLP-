# Word2Vec Model 

```python
from gensim.models import Word2Vec
import pandas as pd
import os
```

```python
data = pd.read_csv('reviews_combined_amazon.csv')
```

```python
data.shape
```

```python
data
```

```python
data['combined_title_text'] = data['review_title'].apply(lambda x:str(x)) + "" + data['review_text'].apply(lambda x:str(x))
```

```python
data['combined_title_text'] 
```

```python
rewiews_title_text_list = data['combined_title_text'].to_list() 
```

```python
rewiews_title_text_list
```

```python
#Gensim accepts a 2D list and hence we gotta convert it to that

sentence_stream = [doc.split(" ") for doc in rewiews_title_text_list]
```

```python
sentence_stream
```

Now, since we split up the sentences into individuals words, some words ,might not make sense, like the word smart phone, the individual words,smart and phone do not make any sense in this context and hence we want those words to be kept kept together. These type of words are called as phrases and we have a model in gensim calld phrases that collects such words from the word stream and then stores it in the same list.

```python
from gensim.models import Phrases
from gensim.models.phrases import Phraser
```

```python
bigram = Phrases(sentence_stream, min_count= 2, threshold=5, delimiter=' ' )
bigram_phraser = Phraser(bigram) #it is not mandatory but it is done for better memeory utlization.
```

```python
token_list = []
for sent in sentence_stream:
    tokens_ = bigram_phraser[sent]
    token_list.append(tokens_)
```

```python
token_list
```

```python
model = Word2Vec(token_list, vector_size=200, window = 5, sg=0)
```

```python
model.wv.most_similar('image')
```

```python
model.wv.save_word2vec_format("word2vec_shopping.txt")
```

#now letus convert this model into something that our model understands

#rasa understands a model as a zip file and hence we will zip it and send


```python
import gzip
fp = open('word2vec_shopping.txt','rb')
data = fp.read()
bindata = bytearray(data)
with gzip.open('word2vec_shopping.txt.gz', 'wb') as f:
    f.write(bindata)
```
