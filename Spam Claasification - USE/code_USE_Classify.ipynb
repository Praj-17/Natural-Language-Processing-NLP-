```python
import tensorflow as tf
import tensorflow_hub as tensorflow_hub
import pandas as pd
from sklearn import preprocessing
import keras
import numpy as np
```

```python
data = pd.read_csv("E:\DATA SETS\spam.csv",encoding = 'latin-1')
```

```python
url = "https://tfhub.dev/google/universal-sentence-encoder-large/4"
embed = tensorflow_hub.load(url)
```

```python
y = list(data['v1'])
x = list(data['v2'])
```

```python
encoder = preprocessing.LabelEncoder()
encoder.fit(y)
```

```python
from keras import utils as np_utils
def encode(encoder,labels):
    encode = encoder.transform(labels)
    return keras.utils.np_utils.to_categorical(encode)
def decode(encoder,one_hot):
    decode = np.argmax(one_hot,axis =1)
    return encoder.inverse_transform(decode)
    
```

```python
test = encode(encoder, ['ham'])
test
```

```python
untest = decode(encoder,test)
untest
```

```python
x_encoded = x
y_encoded = encode(encoder,y)

x_train = np.asarray(x_encoded[:5000])
y_train = np.asarray(y_encoded[:5000])

x_test = np.asarray(x_encoded[5000:])
x_test = np.asarray(y_encoded[5000:])
```

```python
from keras.layers import Input, Lambda, Dense
from keras.models import Model
import keras.backend as K  
```

```python
def UniversalEmbedding(x):
    return embed(tf.squeeze(tf.cast(x,tf.string))) 
```

```python
input_text = Input(shape =(1,), dtype =tf.string)
embedding = Lambda(UniversalEmbedding,output_shape = (512,)) (input_text)
dense = Dense(256,activation = 'relu')(embedding)
pred = Dense(2,activation = 'softmax')(dense)
model = Model(inputs = [input_text], outputs =pred)
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics = ['accuracy'])
```
